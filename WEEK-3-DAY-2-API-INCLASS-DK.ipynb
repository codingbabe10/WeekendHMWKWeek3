# working with APis

<p> What exactly is an API? <br> <br>
API is the acronym for Application Programming Interface, which is a software intermediary that allows two applications to talk to each other. Each time you use an app like Facebook, send an instant message, or check the weather on your phone, you're using an API. </p>

### The Poke API  allows you to retreive a pokemon's information from PokeAPI https://pokeapi.co/



```python
# making an API call
import requests

pokemon = input("What pokemon would you like to search?: ").lower()

response = requests.get(f'https://pokeapi.co/api/v2/pokemon/{pokemon}')


if response.status_code == 200:
    print('success')
    data = response.json()
    print(data)
else:
    print('Error')
```

```python

class Pokemon():
    def __init__(self, name):
        self.name = name
        self.ability = None
        self.type = None
        self.weight = None
        self.hp = None
        self.sprite = None
        self.poke_api_call()

    def poke_api_call(self):
        response = requests.get(f'https://pokeapi.co/api/v2/pokemon/{self.name}')

        if response.status_code == 200:
            data = response.json()
            self.abilitiy = [ability['ability']['name'] for ability in data['abilities']]
            self.type = data['types'][0]['type']['name']
            self.weight = data['weight']
            self.hp = data['stats'][0]['base_stat']
            self.sprite = data['sprites']['front_default']

        else:
            print('Invalid Pokemon Name')

    def evolve(self):
        new_pokemon = input(f"what pokemon do you want {self.name} to evolve into")
        self.name = new_pokemon
        self.poke_api_call()



pokemon = Pokemon('mew')
print(pokemon.name)
print(pokemon.type)
print(pokemon.abilitiy)

pokemon.evolve()


print(pokemon.name)
print(pokemon.type)
print(pokemon.abilitiy)



```
